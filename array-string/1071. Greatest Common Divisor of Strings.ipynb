For two strings `s` and `t`, we say "`t` divides `s`" if and only if `s = t + t + t + ... + t + t` (i.e., `t` is concatenated with itself one or more times).


Given two strings `str1` and `str2`, return the largest string `x` such that `x` divides both `str1` and `str2`.



Example 1:
```
Input: str1 = "ABCABC", str2 = "ABC"

Output: "ABC"
```
Example 2:
```
Input: str1 = "ABABAB", str2 = "ABAB"

Output: "AB"
```
Example 3:
```
Input: str1 = "LEET", str2 = "CODE"

Output: ""
```
Example 4:
```
Input: str1 = "AAAAAB", str2 = "AAA"

Output: ""​​​​​​​
```


Constraints:

```1 <= str1.length, str2.length <= 1000```

`str1` and `str2` consist of English uppercase letters.

````python
class Solution:
	@staticmethod
	def gcdOfStrings(str1: str, str2: str) -> str:
		big_str = str1 if len(str1) >= len(str2) else str2
		small_str = str2 if len(str1) >= len(str2) else str1

		len1 = max(len(str1), len(small_str))
		len2 = min(len(str1), len(small_str))
		multiplo = 1

		# Calcula el divisor mínimo del str2.
		for i in range(len2-1, 1, -1):
			if len2 % i != 0:
				break

			multiplo = int(len2 / (len2 // i))
			# Verifica que el múltiplo obtenido sea correcto y que el patrón existe.
			if small_str[:multiplo] * multiplo == small_str:
				small_str = small_str[:multiplo]
				break

		if multiplo == 1:
			return small_str if small_str * (len1 // len2) == big_str else ""

		# Si `len2` no es par, la retorna como MCD.
		mcd = int(len1 / (len1 // multiplo))
		return small_str if big_str[:mcd] == small_str else ""
````

````python
class Solution:
	@staticmethod
	def gcdOfStrings(str1: str, str2: str) -> str:
		len1 = len(str1)
		len2 = len(str2)
		mcd: int = int()

		for i in range(min(len1, len2), 0, -1):
			if len1 % i == 0 and len2 % i == 0:
				mcd = i
				break

		if str1[:mcd] * (len2 // mcd) == str2 and str2[:mcd] * (len1 // mcd) == str1:
			return str1[:mcd]

		return ""
````

````python
# Best time.
class Solution:
	def gcdOfStrings(self, str1: str, str2: str) -> str:
		if str1 + str2 != str2 + str1:
			return ""

		def gcd(len1, len2):
			while len2:
				len1, len2 = len2, len1 % len2
			return len1

		return str1[:gcd(len(str1), len(str2))]
````
